```python
import os
import numpy as np
import wandb
import h5py
import matplotlib.pyplot as plt
```

```python
api = wandb.Api()
runs = api.runs('ipis-mjkim/eeg-analysis')

for run in runs:
    # print(run, run.config['model'])
    model_name = run.config['model']
    
    if model_name.startswith('[2D]'):
        model_name = '2D-' + model_name[5:]
    elif model_name.startswith('[1D]'):
        model_name = '1D-' + model_name[5:]
    
    run.config['model'] = model_name
    run.config.update()
    run.update()
        
    print(model_name)
```
