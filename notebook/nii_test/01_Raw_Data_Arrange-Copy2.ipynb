```python
import os
import glob
import json
import dicom2nifti
import nibabel as nib
from tqdm.auto import tqdm
from IPython import display
from pprint import pprint
import pandas as pd
```

```python
data_dict = {}
```

```python
with open(r"H:\Other_DB\CAU_EEG\201020\annotation_debug.json") as f:
    anno = json.load(f)
    for d in anno['data']:
        name = d['edfname'].split('_')[0]
        data_dict[name] = data_dict.get(name, set())
        data_dict[name].add('eeg')
```

```python
for f in glob.glob(r'H:\Other_DB\CAU_EEG\231114_3DT1\processed_raw\*.nii'):
    name = os.path.splitext(os.path.basename(f))[0]
    name = name.split('_')[0].split(' ')[0]
    data_dict[name] = data_dict.get(name, set())
    data_dict[name].add('raw')
```

```python
for f in glob.glob(r'H:\Other_DB\CAU_EEG\231114_3DT1\mri\mwp1*.nii'):
    name = os.path.splitext(os.path.basename(f))[0]
    name = name.split('mwp1')[1]
    name = name.split('_')[0].split(' ')[0]
    data_dict[name] = data_dict.get(name, set())
    data_dict[name].add('mwp1')
```

```python
for f in glob.glob(r'H:\Other_DB\CAU_EEG\231114_3DT1\mri\mwp2*.nii'):
    name = os.path.splitext(os.path.basename(f))[0]
    name = name.split('mwp2')[1]
    name = name.split('_')[0].split(' ')[0]
    data_dict[name] = data_dict.get(name, set())
    data_dict[name].add('mwp2')
```

```python
for f in glob.glob(r'H:\Other_DB\CAU_EEG\231114_3DT1\mri\wm*.nii'):
    name = os.path.splitext(os.path.basename(f))[0]
    name = name.split('wm')[1]
    name = name.split('_')[0].split(' ')[0]
    data_dict[name] = data_dict.get(name, set())
    data_dict[name].add('wm')
```

```python
for f in glob.glob(r'H:\Other_DB\CAU_EEG\231114_3DT1\mri\p0*.nii'):
    name = os.path.splitext(os.path.basename(f))[0]
    name = name.split('p0')[1]
    name = name.split('_')[0].split(' ')[0]
    data_dict[name] = data_dict.get(name, set())
    data_dict[name].add('p0')
```

```python
for f in glob.glob(r'H:\Other_DB\CAU_EEG\231114_3DT1\mri\y_*.nii'):
    name = os.path.splitext(os.path.basename(f))[0]
    name = name.split('y_')[1]
    name = name.split('_')[0].split(' ')[0]
    data_dict[name] = data_dict.get(name, set())
    data_dict[name].add('y')
```

```python
# DataFrame 생성을 위한 초기 데이터 구조 설정
columns = ['Name', 'EEG', 'RAW', 'MWP1', 'MWP2', 'WM', 'P0', 'Y']
df_data = {col: [] for col in columns}
df_data['Name'] = list(data_dict.keys())

# 속성 존재 여부에 따른 'O' 또는 'X' 값 할당
for key in data_dict:
    df_data['EEG'].append('O' if 'eeg' in data_dict[key] else '')
    df_data['RAW'].append('O' if 'raw' in data_dict[key] else '')
    df_data['MWP1'].append('O' if 'mwp1' in data_dict[key] else '')
    df_data['MWP2'].append('O' if 'mwp2' in data_dict[key] else '')
    df_data['WM'].append('O' if 'wm' in data_dict[key] else '')    
    df_data['P0'].append('O' if 'p0' in data_dict[key] else '')
    df_data['Y'].append('O' if 'y' in data_dict[key] else '')

# DataFrame 생성
df = pd.DataFrame(df_data)
df = df.sort_values(by='Name').reset_index(drop=True)

df.head()  # 처음 몇 줄을 출력하여 DataFrame을 확인
```
